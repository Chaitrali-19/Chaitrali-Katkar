```python
plt.figure(figsize=(8,8))
df['Sub-Category'].value_counts().plot.pie(autopct='dark')
plt.show()
```

```python
sns.distplot(df.Sales)
```

```python
sns.distplot(df.Profit)
```

```python
sns.distplot(df.Discount)
```

```python
sns.distplot(df.Quantity)
```

```python
df1=df['City'].value_counts()
df1=df1.head(100)
```

```python
df1.plot(kind='bar',figsize=(20,15))
plt.xlabel('City')
plt.ylabel('count')
plt.show()
```

```python
df.plot(x='Quantity',y='Sales',style='o')
plt.title('Quantity vs Sales')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.show()
```

```python
df.plot(x='Discount',y='Profit',style='o')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```
